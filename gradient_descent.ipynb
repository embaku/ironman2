## Gradient Descent 

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.arange(-20,10)
Y = (X+5)**2

print(X)
print(Y)
```

## Visuallization

```python
plt.style.use('seaborn')
plt.plot(X,Y)
plt.ylabel('F(X)')
plt.xlabel('X')
plt.show()
```

```python
x = 3
lr = 0.01
steps = 2000
precision = 0.000001
```

```python
for i in range(steps):
    
    grad = 2*(x+5)
    prev_x = x
    x = x- lr*grad
    
    print("Iteration: ",i+1)
    print("local minima occurs at: ",x)
    
    delta_x = abs(prev_x - x)
    
    if delta_x <= precision:
        break
```
